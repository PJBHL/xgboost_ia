Pré-processamento:
- Retirada do atributo unnamed (inútil para as analises).
- Retirada de dada inicial e final (redundantes com mês e ano).
- Pegando somente gasolina, desconsiderando outras instâncias de outros combustíveis.

```python
df.drop(['Unnamed: 0', 'UNIDADE DE MEDIDA'], axis = 'columns', inplace = True)
```

```python
df.PRODUTO.unique().tolist()
```

```python
df.groupby('PRODUTO')
df_etanol = df.loc[df.PRODUTO =='ETANOL HIDRATADO']
df_gasolina = df.loc[df.PRODUTO =='GASOLINA COMUM']
df_glp = df.loc[df.PRODUTO =='GLP']
df_gnv = df.loc[df.PRODUTO == 'GNV']
df_diesel = df.loc[df.PRODUTO == 'ÓLEO DIESEL']
df_diesel_s10 = df.loc[df.PRODUTO == 'ÓLEO DIESEL S10']
```

"Valores Ausentes"

A base tem alguns valores faltantes que tem "-", como o xgboost trabalha com dados ausentes basta substituir "-" por um vazio "".

```python
missingValues = ['MARGEM MÉDIA REVENDA','DESVIO PADRÃO DISTRIBUIÇÃO','PREÇO MÍNIMO DISTRIBUIÇÃO','PREÇO MÁXIMO DISTRIBUIÇÃO','COEF DE VARIAÇÃO DISTRIBUIÇÃO']
```

```python
for i in missingValues:
    df_gasolina[i] = pd.to_numeric(df_gasolina[i].str.replace('-',''))
```

- Descartando atributos 'DATA INICIAL' e 'DATA FINAL' -> são redundantes com mês e ano.

```python
df_gasolina.drop(['DATA INICIAL','DATA FINAL'],axis=1,inplace=True)
```

```python
df_gasolina.head(10)
```

```python
df_gasolina1 = df_gasolina[['ANO','MÊS']]
```

## Import do xgboost Python

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error,mean_squared_log_error,r2_score,explained_variance_score
```

```python
y = df_gasolina['PREÇO MÉDIO REVENDA']
```

## Conjuntos de Treino:

```python
x_treino,x_teste,y_treino,y_teste = train_test_split(df_gasolina1,y,test_size=0.15,random_state=23)
```

```python
xgbr= XGBRegressor()
xgbr.fit(x_treino,y_treino)
```

```python
xgbr.score(x_teste,y_teste)
```

- GridSearch para encontrar os melhores parâmetros.

```python
xgbr1= XGBRegressor()
parameters = {'objective':['reg:linear'],
              'learning_rate': [.1, 0.2, .5,0.01],
              'max_depth': [2,3],
              'min_child_weight': [0.5,1],
              'subsample': [0.5,1],
              'colsample_bytree': [0.5,1],
              'n_estimators': [100,200,300]}
xgb_grid = GridSearchCV(xgbr1,
                        parameters,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)
xgb_grid.fit(x_treino,y_treino)
```

- Melhores parâmetros encontrados:

```python
xgb_grid.best_params_
```

```python
improved_xgbr =XGBRegressor(colsample_bytree= 1, learning_rate= 0.41, max_depth= 3,
                          min_child_weight = 0.5, n_estimators = 333, objective = 'reg:linear', subsample=1)
improved_xgbr.fit(x_treino,y_treino)
```

```python
improved_pred = improved_xgbr.predict(x_teste)
improved_xgbr.score(x_teste,y_teste)
```

- Métricas de Avaliação para o modelo:

```python
print(mean_squared_error(y_teste,improved_pred))
print(mean_squared_log_error(y_teste,improved_pred))
print(r2_score(y_teste,improved_pred))
print(explained_variance_score(y_teste,improved_pred))
```

- Teste de previsões, dado data e mês.

```python
teste= [(2019,6),(2019,7),(2019,8),(2019,9),(2019,10),(2019,11),(2019,12)]
dft = pd.DataFrame(teste,columns=['ANO','MÊS'])
improved_xgbr.predict(dft)
```
